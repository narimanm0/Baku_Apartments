```python
import pandas as pd
from config.config import CLEANED_CSV_FILE

df = pd.read_csv(CLEANED_CSV_FILE)
df
```

```python
df["Category"].unique()
```

```python
category_to_file = {
    'Yeni tikili': 'data/yeni_tikili_data.csv',
    'Torpaq': 'data/torpaq_data.csv',
    'Həyət evi/Bağ evi': 'data/heyet_evi_data.csv',
    'Köhnə tikili': 'data/kohne_tikili_data.csv',
    'Obyekt': 'data/obyekt_data.csv',
    'Ofis': 'data/ofis_data.csv',
    'Qaraj': 'data/qaraj_data.csv'
}

for category, filename in category_to_file.items():
    category_df = df[df['Category'] == category]
    if not category_df.empty:
        category_df.to_csv(filename, index=False)
        print(f"Saved {len(category_df)} records to {filename}")
    else:
        print(f"No data found for category: {category}, skipping...")
```
