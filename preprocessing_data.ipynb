```python
import pandas as pd
from config.config import OUTPUT_CSV_FILE, CLEANED_CSV_FILE
```

```python
df = pd.read_csv(OUTPUT_CSV_FILE, index_col=0) # index_col=0 for removing meaningless ID
df.drop(columns=['Binanın növü'], inplace=True)
df
```

```python
column_mapping = {
    'item_id': 'Item_id',
    'amount': 'Amount',
    'currency': 'Currency',
    'location': 'Location',
    'Kateqoriya': 'Category',
    'Sahə': 'Area',
    'Torpaq sahəsi': 'Land_area',
    'Otaq sayı': 'Number_of_rooms',
    'Çıxarış': 'Frontage',
    'İpoteka': 'Mortgage',
    'Təmir': 'Condition',
    'Mərtəbə': 'Floor',
}

df.rename(columns=column_mapping, inplace=True)
df.columns
```

```python
columns_to_update = ['Frontage', 'Mortgage', 'Condition']

for col in columns_to_update:
    df[col] = df[col].replace({'var': 1, 'yoxdur': 0})
    df[col] = df[col].fillna(0).astype(int)
df
```

```python
df[['Current_floor', 'Total_floors']] = df['Floor'].str.split(' / ', expand=True)

df['Current_floor'] = pd.to_numeric(df['Current_floor'], errors='coerce')
df['Total_floors'] = pd.to_numeric(df['Total_floors'], errors='coerce')
df.drop(columns=['Floor'], inplace=True)

df
```

```python
df['Land_area'] = df['Land_area'].astype(str)

mask = df['Area'].str.contains('sot', case=False, na=False)

sot_values = df.loc[mask, 'Area']

df.loc[mask, 'Land_area'] = sot_values
df.loc[mask, 'Area'] = '0'

df['Land_area'] = df['Land_area'].replace('nan', '0').fillna('0')

df
```

```python
df['Number_of_rooms'] = df['Number_of_rooms'].fillna(0)
df['Current_floor'] = df['Current_floor'].fillna(0)
df['Total_floors'] = df['Total_floors'].fillna(0)
df.to_csv(CLEANED_CSV_FILE)
print(df.head(10))
print(df.describe)
```
